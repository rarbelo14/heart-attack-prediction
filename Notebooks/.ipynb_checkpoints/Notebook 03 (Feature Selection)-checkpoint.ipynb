# 1. Importación de paquetes

```python
# Para la lectura y transformación de datos
import pandas as pd

# Para el manejo de vectores
import numpy as np

# Para la visualización de datos
import matplotlib.pyplot as plt

# Modelo de regresión logística 
from sklearn.linear_model import LogisticRegression

# Para la selección de variables mediante Recursive Feature Elimination
from sklearn.feature_selection import RFE

# Para realizar validación cruzada repetida
from sklearn.model_selection import RepeatedKFold

# Para conocer el error por validación cruzada
from sklearn.model_selection import cross_val_score

# Para evaluar las predicciones
from sklearn.metrics import roc_auc_score

# Modelo de Random Forest
from sklearn.ensemble import RandomForestClassifier
```

# 2. Lectura de los datasets procedentes del Notebook 2

```python
df_train = pd.read_csv(r'../Dataset_con_transformaciones/df_train_after_notebook2.csv')
df_test = pd.read_csv(r'../Dataset_con_transformaciones/df_test_after_notebook2.csv')
```

Dividir los conjuntos de train y test en:
- Variables input 
- Variable objetivo

```python
X_train, y_train = df_train.drop(columns=['output']), df_train.output
X_test, y_test = df_test.drop(columns=['output']), df_test.output
```

# 3. Feature Selection

A la hora de realizar la selección de variables con el método stepwise, y después de un estudio de distintos paquetes de Python y funciones de scikit-learn, se ha llegado a la conclusión de en este lenguaje de programación no existe tal método. Por ende, se opta por usar la función RFE (Recursive Feature Elimination).

https://datascience.stackexchange.com/questions/937/does-scikit-learn-have-a-forward-selection-stepwise-regression-algorithm

## 3.1. RFE con regresión logística

Para hacer validación cruzada repetida, usar RepeatedKFold:

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RepeatedKFold.html#sklearn.model_selection.RepeatedKFold

**¿Por qué escoger RepeatedKFold para validación cruzada repetida (y no elegir RepeatedStratifiedKFold)?**

Como los datos no están desbalanceados, no hace falta utilizar RepeatedStratifiedKFold. 

https://medium.com/@venkatasujit272/overview-of-cross-validation-3785d5414ece

**¿Qué scoring escoger?**

https://scikit-learn.org/stable/modules/model_evaluation.html#scoring

Como se vio en el módulo "Minería de datos y modelización predictiva - Aida Calviño", para problemas de clasificación, la técnica del área bajo la curva ROC (AUROC) es una medida bastante acertada a la hora de evaluar este tipo de algoritmos predictivos.

```python
num_vars = []
cv_auroc_scores = []
cv_auroc_scores_mean = []
cv_auroc_scores_median = []
pred_auroc_scores = []

cv = RepeatedKFold(n_splits=5, n_repeats=10, random_state=99) # Validación cruzada repetida

for n in list(range(3, X_train.shape[1] + 1)): # Rango de variables de 3 a 28 (número total de variables de entrada)
    num_vars.append(n)
    
    # Modelo con n variables de entrada
    model = LogisticRegression(random_state=99)
    rfe = RFE(estimator=model, n_features_to_select=n)
    rfe.fit(X_train, y_train)

    # ¿Cuáles son las variables seleccionadas?
    selected_features = X_train.columns[rfe.support_]

    # Accuracy en el conjunto de entrenamiento
    cv_auroc = cross_val_score(model, X_train[selected_features], y_train, cv=cv, scoring='roc_auc')
    cv_auroc_mean = np.mean(cv_auroc)
    cv_auroc_median = np.median(cv_auroc)
    
    cv_auroc_scores.append(cv_auroc)
    cv_auroc_scores_mean.append(cv_auroc_mean)
    cv_auroc_scores_median.append(cv_auroc_median)
    
    # Accuracy en el conjunto de evaluación
    model.fit(X_train[selected_features], y_train)
    y_pred = model.predict(X_test[selected_features])
    pred_auroc = roc_auc_score(y_test, y_pred)
    
    pred_auroc_scores.append(pred_auroc)
    
    print(f"Número de variables: { n }")
    print(f"Media del área bajo la curva ROC obtenido en validación cruzada (cross_val_score): { round(cv_auroc_mean, 3) }")
    print(f"Mediana del área bajo la curva ROC obtenido en validación cruzada (cross_val_score): { round(cv_auroc_median, 3) }")
    print(f"Área bajo la curva ROC obtenido en las predicciones: {round(pred_auroc, 3)}")
    print(f"Variables seleccionadas: { list(selected_features) }")
    print("\n---\n")
```

```python
plt.figure(figsize=(20, 30))

plt.subplot(3, 1, 1)
# Repetir la cantidad de variables 50 veces (una cantidad de variables por cada split y repetición en el CV)
repeated_num_vars = np.repeat(num_vars, 50) 
cv_auroc_scores_flatten = [ auroc for auroc_list in cv_auroc_scores for auroc in auroc_list ]
results_df = pd.DataFrame(list(zip(repeated_num_vars, cv_auroc_scores_flatten)), columns=['num_vars', 'roc_auc'])
grouped_data = results_df.groupby('num_vars')['roc_auc'].apply(list)
plt.boxplot(grouped_data.values, positions=range(1, len(grouped_data) + 1))
sorted_index = sorted(grouped_data.index)  # Ordenar los índices de forma ascendente
plt.xticks(range(1, len(grouped_data) + 1), sorted_index, rotation=90)
plt.ylabel('ROC AUC')
plt.title('ROC AUC por número de variables en validación cruzada')

plt.subplot(3, 1, 2)
plt.plot(num_vars, cv_auroc_scores_mean, color='blue', label='Media') # Graficar la media
plt.plot(num_vars, cv_auroc_scores_median, color='green', label='Mediana') # Graficar la mediana
plt.ylabel('ROC AUC')
plt.title('ROC AUC por número de variables en validación cruzada')
plt.legend()
plt.xticks(num_vars)
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(num_vars, np.array(pred_auroc_scores))
plt.ylabel('ROC AUC')
plt.title('ROC AUC por número de variables en predicciones con el test-set')
plt.xticks(num_vars)
plt.grid(True)

plt.show()
```

Según los resultados del área bajo la curva ROC y los resultados de precisión obtenidos en la validación cruzada con los conjuntos de entrenamiento, teniendo también en consideración el área bajo la curva ROC de las predicciones obtenidas con los conjuntos de validación y, por último, no dejando de lado el principio de parsimonia (el cual se refiere a preferir modelos más simples y con menos parámetros cuando se enfrenta a opciones de igual rendimiento predictivo), se observa que generar modelos con 11 variables parece ser una opción favorable.

En los resultados proporcionados, se puede observar que el modelo con 11 variables seleccionadas obtiene un área bajo la curva ROC promedio en la validación cruzada (cross_val_score) de 0.935 y un área bajo la curva ROC en las predicciones de 0.781. 

Por un lado, estas cifras son superiores o comparables a las obtenidas con un mayor números de variables, mientras que el modelo con 11 variables mantiene la simplicidad y la capacidad de generalización del modelo. Por otro lado, el modelo generado con 6 variables presenta también muy buenos resultados del área bajo la curva ROC, pero su presición es algo más baja. Por ello se ha optado por escoger el modelo con 11 variables.

## 3.2. Parámetro "feature_importances_" en un modelo Random Forest

Esto es un método alternativo a RFE con regresión logística

```python
model = RandomForestClassifier(random_state=99)
model.fit(X_train, y_train)

importances = model.feature_importances_

# Visualización de la importancia de las variables
sorted_indices = np.argsort(importances)[::-1]
sorted_importances = importances[sorted_indices]
sorted_labels = X_train.columns[sorted_indices]
plt.figure(figsize=(10, 6))
plt.bar(range(len(sorted_importances)), sorted_importances)
plt.xticks(range(len(sorted_importances)), sorted_labels, rotation=90)
plt.xlabel('Características')
plt.ylabel('Importancia')
plt.title('Importancia de las características')
plt.show()
```

Como se puede observar en el análisis realizado, las 11 variables seleccionadas previamente mediante Recursive Feature Elimination (RFE) (['trtbps', 'chol', 'oldpeak', 'sex_1', 'cp_0', 'exng_1', 'slp_1', 'caa_0.0', 'caa_2.0', 'thall_2', 'thall_3']) presentan una importancia relativa considerable según los resultados obtenidos mediante el algoritmo Random Forest.

En el gráfico superior, se puede apreciar que las variables no seleccionadas coinciden con las variables de menor importancia calculadas por el modelo Random Forest. Esto indica que las variables seleccionadas previamente se alinean con las variables de mayor importancia según el algoritmo Random Forest, lo que respalda y justifica la elección de estas.
