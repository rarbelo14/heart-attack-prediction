## 3.2. Estudio de la unión entre variables

### 3.2.1. Variables categóricas con otras categóricas

```python
# Chi cuadrado

relation_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols)

for i in range(len(cat_cols)):
    for j in range(i+1, len(cat_cols)):
        contingency_table = pd.crosstab(df[cat_cols[i]], df[cat_cols[j]])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        relation_matrix.at[cat_cols[i], cat_cols[j]] = p
        relation_matrix.at[cat_cols[j], cat_cols[i]] = p

relation_matrix = relation_matrix.astype(float).apply(lambda x: x.round(2))

# Mapa de calor
sns.heatmap(relation_matrix, mask=np.triu(np.ones_like(relation_matrix, dtype=bool)), annot=True)

plt.show()
```

Según la prueba de Chi cuadrado, existe una fuerte relación entre:
- "fbs" y "sex"
- "restecg" y "sex"
- "slp" y "sex"
- "fbs" y "exng"
- "restecg" y "thall"

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()
axes[5].axis('off') # Eliminar el noveno plot

sns.countplot(data=df, x="fbs", hue="sex", ax=axes[0])
sns.countplot(data=df, x="restecg", hue="sex", ax=axes[1])
sns.countplot(data=df, x="slp", hue="sex", ax=axes[2])
sns.countplot(data=df, x="fbs", hue="exng", ax=axes[3])
sns.countplot(data=df, x="restecg", hue="thall", ax=axes[4])

plt.show()
```

Observamos que, mediante la prueba de Chi cuadrado, no se obtienen conclusiones determinantes a la hora de analizar las relaciones entre variables categóricas.

Se opta por probar la técnica de V de Cramer con el objetivo de obtener conclusiones más fiables acerca de la correlación entre las variables categóricas

```python
# V de Cramer

def cramers_v(contingency_table):
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    phi2 = chi2 / n
    r, k = contingency_table.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

relation_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols)

for i in range(len(cat_cols)):
    for j in range(i + 1, len(cat_cols)):
        contingency_table = pd.crosstab(df[cat_cols[i]], df[cat_cols[j]])
        cramer_v = cramers_v(contingency_table)
        relation_matrix.at[cat_cols[i], cat_cols[j]] = cramer_v
        relation_matrix.at[cat_cols[j], cat_cols[i]] = cramer_v

relation_matrix = relation_matrix.astype(float).apply(lambda x: x.round(2))

# Mapa de calor
sns.heatmap(relation_matrix, mask=np.triu(np.ones_like(relation_matrix, dtype=bool)), annot=True)

plt.show()
```

Existe una fuerte relación entre **"exng"** y **"cp"**. El resto de variables categóricas parecen no tener relación las unas con las otras.

```python
sns.countplot(data=df, x="exng", hue="cp")
plt.show()
```

Mediante la prueba de V de Cramer, podemos concluir lo siguiente:
- En aquellos sujetos con una angina no inducida por el ejercicio o sin una angina inducida por el ejercicio (exng = 0), muchos de estos casos presentan un dolor no anginoso (cp = 2).
- En aquellos sujetos con una angina inducida por el ejercicio (exng = 1), en la mayor parte de los casos estos presentan una angina típica (cp = 0).

A partir del gráfico superior, se observa que existen registros con valores iguales a 2 (dolor no anginoso) o a 3 (asintomático) en la variable **"cp"** que tienen un valor igual a 1 (el usuario tiene una angina inducida por el ejercicio) en la variable **"exng"**. Estos casos podrían ser considerados como errores humanos a la hora de introducir los datos en el sistema sanitario, por lo que deberían de ser eliminados, ya que generarían ruido en los modelos. Entonces, para este proyecto, se ha optado por convertir a valor 0 la variable exng en aquellos registros en donde  la variable cp tenga un valor igual a 2 o a 3.

## 3.3. Estudio de outliers en variables numéricas

De acuerdo con [este blog](https://towardsdatascience.com/detecting-and-treating-outliers-in-python-part-1-4ece5098b755), los datos atípicos se identifican de la siguiente manera cuando la variable no sigue una distribución normal (lo cual sucede en todas las variables numéricas de este dataset):

Límite superior = Tercer cuartil + 1.5 * IQR

Límite inferior = Primer cuartil - 1.5 * IQR

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()
axes[5].axis('off') # Eliminar el sexto plot

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x=col, ax=axes[i])
```

```python
def outliers_analysis(data):
    '''
    Función que calcula los límites para outliers de la serie,
    identifica los valores atípicos y calcula la proporción de estos en la serie.
    
    Input:
        data { Series }: conjunto de valores numéricos de una columna en el dataframe
        
    Output: 
        lower_bound { float }: límite inferior para identificar outliers
        upper_bound { float }: límite superior para identificar outliers
        np.mean(outliers) { float }: porcentaje de outliers presentes en la serie data pasada por parámetro
    '''
    # Calcular el rango intercuartil (IQR)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    # Definir los límites para identificar outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    # Identificar outliers
    outliers = (data < lower_bound) | (data > upper_bound)
    return lower_bound, upper_bound, np.mean(outliers)

for col in num_cols:
    _, _, outliers_percentage = outliers_analysis(df[col])
    print(f"Porcentaje de outliers para la variable \"{col}\" es {(outliers_percentage * 100).round(2)} %")
```

Para la variable **"age"**, no existen registros atípicos.

Para la variable **"trtbps"**, existen registros por fuera de los límites para identificar outliers. Sin embargo, estos registros suponen un 2.97 % del total de registros.
>Podrían no ser considerados como atípicos realmente, al ser su frecuencia relativa mayor al 2 %.
>> Como su propio nombre indica, los datos atípicos son "raros", por lo que necesariamente han de representar una proporción pequeña del conjunto de datos. 
>>
>>El valor 2 % es un valor que, en este proyecto específicamente, se ha tomado para determinar cuándo considerar outliers en una variable y cuando no, debido a que el dataset cuenta con 303 registros unicamente, por lo que no conviene desperdiciar datos. 

El resto de variables sí presentan valores atípicos.
> Eliminar estos valores, ya que no representan una proporción considerable del dataset. Posteriormente, en el tratamiento de datos faltantes, se imputarán valores a estos missings.

```python
# Convertir a null los outliers
for col in ["chol", "thalachh", "oldpeak"]:
    lower_bound, upper_bound, _ = outliers_analysis(df[col])
    df[col] = df[col].apply(lambda x: None if (x < lower_bound) or (x > upper_bound) else x)
```
